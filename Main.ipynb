```python
num_df = df.select_dtypes(include=['int64','float64'])
non_num_df = df.select_dtypes(include=['object'])
```

```python
for i in num_df.columns:
    sns.kdeplot(num_df[i],fill=True)
    plt.title(i)
    plt.show()
```

```python
data_bath = df['bathrooms'].value_counts().reset_index()
data_bath.columns = ['bathroom', 'count']

plt.figure(figsize=(8,6))
sns.barplot(x='bathroom',y='count', data=data_bath)
plt.show()
```

```python
data_bedroom = df['bedrooms'].value_counts().reset_index()
data_bedroom.columns = ['bedrooms', 'count']
plt.figure(figsize=(8,6))
sns.barplot(x='bedrooms', y= 'count',data=data_bedroom)
plt.show()
```

```python
data_floor = df['floors'].value_counts().reset_index()
data_floor.columns = ('floors', 'count')

plt.figure(figsize=(8,6))

sns.barplot(x='floors', y='count',data=data_floor)
plt.show()
```

```python
data_cond = df['condition'].value_counts().reset_index()
data_cond.columns = ['condition', 'count']

plt.figure(figsize=(8,6))

sns.barplot(x='condition', y= 'count',data=data_cond)
plt.show()
```

```python
data_city = df['city'].value_counts().to_frame().reset_index()

data_city.columns = ['city','count']

data_city = data_city.sort_values(ascending=True,by='count')

plt.figure(figsize=(8,10))

plt.barh(data_city['city'], data_city['count'])

plt.show()
```

```python
city_data_per_median = df.groupby('city')['price'].median().to_frame().reset_index()
city_data_per_median=city_data_per_median.sort_values(ascending=False,by='price')[:20]
plt.figure(figsize=(8,12))

sns.barplot(x='price', y='city',data=city_data_per_median)

plt.show()
```
